```python
import os
import pandas as pd

base_path = "data/raw/donateacry_corpus_cleaned_and_updated_data"

data = []

labels = os.listdir(base_path)

for label in labels:
    label_path = os.path.join(base_path, label)
    
    if os.path.isdir(label_path):
        files = os.listdir(label_path)
        
        for file in files:
            file_path = os.path.join(label, file)
            data.append([file_path, label])

df = pd.DataFrame(data, columns=["file_path", "label"])

df.to_csv("data/master_metadata.csv", index=False)

print("Metadata file created!")
```

```python
import numpy as np
import soundfile as sf

# create noise
noise = np.random.randn(len(audio))

# add noise to original audio
augmented_audio = audio + 0.005 * noise

# save new file
output_path = "data/raw/donateacry_corpus_cleaned_and_updated_data/burping/aug_noise.wav"
sf.write(output_path, augmented_audio, sr)

print("Augmented file saved!")
```

```python
import os
import librosa
import numpy as np
import soundfile as sf

raw_base = "data/raw/donateacry_corpus_cleaned_and_updated_data"
aug_base = "data/processed/augmented"

target_classes = ["burping", "belly_pain", "discomfort"]

for label in target_classes:
    raw_path = os.path.join(raw_base, label)
    aug_path = os.path.join(aug_base, label)

    files = os.listdir(raw_path)

    for file in files:
        file_path = os.path.join(raw_path, file)

        audio, sr = librosa.load(file_path, sr=None)

        # aug 1 noise
        noise = np.random.randn(len(audio))
        audio_noise = audio + 0.005 * noise

        new_name1 = "noise_" + file
        sf.write(os.path.join(aug_path, new_name1), audio_noise, sr)

        # aug 2 loudness
        audio_loud = audio * 1.2

        new_name2 = "loud_" + file
        sf.write(os.path.join(aug_path, new_name2), audio_loud, sr)

print("Augmentation complete!")
```

```python
target_classes = ["burping", "tired", "belly_pain"]

for label in target_classes:
    raw_path = os.path.join(raw_base, label)
    aug_path = os.path.join(aug_base, label)

    files = os.listdir(raw_path)

    for file in files:
        file_path = os.path.join(raw_path, file)

        audio, sr = librosa.load(file_path, sr=None)

        for i in range(4):  # stronger boost
            noise = np.random.randn(len(audio))
            audio_noise = audio + 0.005 * noise

            new_name = f"extra_{i}_{file}"
            sf.write(os.path.join(aug_path, new_name), audio_noise, sr)

print("Final augmentation done!")
```
